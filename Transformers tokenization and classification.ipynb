```python
from transformers import pipeline
```

```python
classifier=pipeline('sentiment-analysis')
```

```python
classifier('the retaurent is good but the food is bad')
```

```python
classifier('the center is very informative and good')
```

```python
classifier('the food is average')
```

```python
result=classifier(["the food is great ", "the huggingface platform is average"])
for results in result:
  print(f"label: {results['label']}, with score: {round(results['score'],4)}")
```

```python
# Load model directly
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
model_name='nlptown/bert-base-multilingual-uncased-sentiment'
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)
#model_name='nlptown/bert-base-multilingual-uncased-sentiment

tokenizer = AutoTokenizer.from_pretrained(model_name)
#model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
classifier=pipeline('sentiment-analysis',model=model,tokenizer=tokenizer)
```
