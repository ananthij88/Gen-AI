```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Simulated data: Daily births (simplified example)
dates = pd.date_range(start='2022-12-20', end='2023-01-10')
births = np.random.poisson(100, len(dates)) + 20 * (dates.day == 25)  # Spike on Christmas

# Plotting the data
plt.plot(dates, births, label='Daily Births')

# Find the correct indices for Christmas and New Year
christmas_index = dates.get_loc('2022-12-25')  # Find the index of Christmas
new_year_index = dates.get_loc('2023-01-01')   # Find the index of New Year

# Highlighting Christmas and New Year
plt.annotate('Christmas', xy=('2022-12-25', births[christmas_index]),
             xytext=('2022-12-22', births[christmas_index] + 20),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('New Year', xy=('2023-01-01', births[new_year_index]),
             xytext=('2022-12-29', births[new_year_index] + 20),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Births')
plt.title('Effect of Holidays on US Births')

# Adding the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Loading the CSV file
df = pd.read_csv('sales_data.csv')

# Displaying the first few rows of the dataset
print(df.head())
```

```python
# Converting the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extracting year, month, and day into new columns
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

print(df.head())
```

```python
# Checking for NaN values
print(df.isnull().sum())

# Removing rows with NaN values
df = df.dropna()

# Alternatively, filling NaN values with the mean of the column
df['Profit'] = df['Profit'].fillna(df['Profit'].mean())

print(df.head())
```

```python
# Applying descriptive statistics
summary = df.describe()
print(summary)

# Descriptive statistics for a specific column
profit_summary = df['Profit'].describe()
print(profit_summary)
```

```python
# Renaming columns
df.rename(columns={'Sales': 'Total_Sales', 'Profit': 'Total_Profit'}, inplace=True)

# Changing data types (e.g., converting 'Quantity' to float)
df['Quantity'] = df['Quantity'].astype(float)

print(df.dtypes)
```

```python
# Dropping unnecessary columns
df = df.drop(columns=['Year', 'Month', 'Day'])

print(df.head())
```

```python
# Assuming the 'Date' column is in UTC and needs to be converted to 'US/Eastern'
df['Date'] = df['Date'].dt.tz_localize('UTC').dt.tz_convert('US/Eastern')

print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Create grid and multivariate normal
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Contour plot
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(label='Function Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Create grid and multivariate normal
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Filled contour plot
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='Function Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filled Contour Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a map
m = Basemap(projection='merc', llcrnrlat=32, urcrnrlat=42, llcrnrlon=-125, urcrnrlon=-114, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='lightgreen', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')

# City coordinates
cities = {'Los Angeles': (34.05, -118.25),
          'San Francisco': (37.77, -122.42),
          'San Diego': (32.72, -117.16),
          'Sacramento': (38.58, -121.49)}

# Plot cities
for city, (lat, lon) in cities.items():
    x, y = m(lon, lat)
    m.plot(x, y, 'bo', markersize=8)
    plt.text(x, y, city, fontsize=12, ha='right')

plt.title('Major Cities in California')
plt.show()
```
